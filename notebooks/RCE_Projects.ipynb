```python
import ipywidgets as widgets
from IPython.display import display

# Replace this with your actual list of techniques
techniques = ["Technique1", "Technique2", "Technique3", "Technique4"]

# Create a top-level tab widget
top_tab = widgets.Tab()

# Create a dictionary to store child tabs
child_tabs = {}

# Iterate through the techniques and create child tabs
for index, technique in enumerate(techniques):
    # Create a child tab for each technique
    child_tab = widgets.Tab()
    
    # Set the title of the child tab to the technique name
    child_tab.set_title(0, technique)
    
    # Create content for the child tab (you can customize this part)
    content = widgets.Text(value=f"This is the content for {technique}")
    
    # Add the content to the child tab
    child_tab.children = [content]
    
    # Add the child tab to the dictionary
    child_tabs[technique] = child_tab
    
    # Add the child tab to the top-level tab
    top_tab.set_title(index, technique)
    top_tab.children = list(child_tabs.values())

# Display the top-level tab widget
display(top_tab)
```
